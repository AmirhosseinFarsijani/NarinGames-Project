```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 16))

fig = sns.barplot(x="score", y='downloads', data=games, ax=axes[0], ci=None)
fig.set(xlabel='scores', ylabel='average of downloads')

fig = sns.barplot(x="score", y='score_num', data=games, ax=axes[1], ci=None)
fig.set(xlabel='scores', ylabel='average number of vote')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

fig1 = sns.lineplot(x='score', y='downloads', data=games, ax=ax, ci=None, color='r',label='average of downloads')
fig1.set(xlabel='scores', ylabel='average of downloads')

ax2 = ax.twinx()
fig2 = sns.lineplot(x='score', y='score_num', data=games, ax=ax2, ci=None, color='b',label='average number of vote')
fig2.set(xlabel='scores', ylabel='average number of vote')

ax.legend(loc='upper left', fontsize='large')
ax2.legend(fontsize='large')

plt.show()
```
